```python
# ANS 1.

import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from collections import Counter


nltk.download('punkt')
nltk.download('stopwords')

text = """Basketball is one of the most exciting sports in the world. 
It requires speed, teamwork, and strategy to win games. 
I love watching NBA matches, especially when the score is close in the final minutes. 
My favorite player is known for his amazing dunks and clutch performances. 
Playing basketball with friends is a great way to stay active and improve coordination."""

print("Ans 1.")

text_clean = re.sub(r'[^\w\s]', '', text.lower())
print("\n(i)",text_clean)

sent_tokens = sent_tokenize(text)
word_tokens_nltk = word_tokenize(text_clean)
print("\n(ii) Words-> ",word_tokens_nltk)
print("(ii) Sentence -> " ,sent_tokens )

split_tokens = text_clean.split()

print("\n(iii) split() -> \n")
print(split_tokens)

print("\n (iii) word_tokenize()-> \n")
print(word_tokens_nltk)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokens_nltk if word not in stop_words]

word_freq = Counter(filtered_words)


print("\n(iv) ",filtered_words)

print("\n(v) Word Frequency Distribution (excluding stopwords):")
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
# ANS 2.

from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')


print("Ans 2.")


alpha_words = re.findall(r'\b[a-zA-Z]+\b', text.lower())

print("\n(i) ",alpha_words)


stop_words = set(stopwords.words('english'))
filtered_words = [word for word in alpha_words if word not in stop_words]

print("\n(ii) ",filtered_words)


stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\n(iii) " ,stemmed_words)


lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\n(iv) ",lemmatized_words)


print("\n(v) Comparison: \n")

words = ["performances", "watching", "playing", "coordination"]
print(f"{'Word':<15}{'Stemmed':<15}{'Lemmatized':<15}")
print("-" * 45)

for word in words:
    stemmed = stemmer.stem(word)
    lemmatized = lemmatizer.lemmatize(word)
    print(f"{word:<15}{stemmed:<15}{lemmatized:<15}")
```

```python
# ANS 3.


from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np


texts = [
    "Scientists discover a new species of dinosaur in Argentina.",
    "Global markets surge as tech stocks rebound after major dip.",
    "Heavy rainfall causes flooding in several parts of the city."
]
print("Ans 3.")

count_vec = CountVectorizer()
bow_matrix = count_vec.fit_transform(texts)

print("\n\n(i)")
print(count_vec.get_feature_names_out())
print(bow_matrix.toarray())


tfidf_vec = TfidfVectorizer()
tfidf_matrix = tfidf_vec.fit_transform(texts)
feature_names = tfidf_vec.get_feature_names_out()
tfidf_array = tfidf_matrix.toarray()

print("\n\n(ii)")
print(feature_names)
print(tfidf_array)


print("\n\n(iii)")
for i, row in enumerate(tfidf_array):
    top_indices = row.argsort()[-3:][::-1]  
    top_keywords = [(feature_names[j], row[j]) for j in top_indices]
    print(f"\nText {i+1}:")
    for word, score in top_keywords:
        print(f"{word}: {score:.4f}")
```

```python
# ANS 5.
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt


reviews = [
    "This laptop has excellent battery life and performance.",
    "The delivery was delayed and the packaging was damaged.",
    "It's okay, not great but not bad either.",
    "The customer service was friendly and quick to resolve my issue.",
    "The product stopped working after a week. Very disappointing."
]


results = []

for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    

    if polarity > 0.1:
        sentiment = "Positive"
    elif polarity < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
        
    results.append({
        "review": review,
        "polarity": polarity,
        "subjectivity": subjectivity,
        "sentiment": sentiment
    })

print("ANS 5.\n")
print("\n(ii)")
for res in results:
    print(f"\nReview: {res['review']}")
    print(f"Polarity: {res['polarity']:.2f}, Subjectivity: {res['subjectivity']:.2f}")
    print(f"Sentiment: {res['sentiment']}")


positive_text = " ".join(res["review"] for res in results if res["sentiment"] == "Positive")

if positive_text:
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_text)


    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title("\n(iii) Word Cloud for Positive Reviews")
    plt.show()
else:
    print("\n(iii) No positive reviews found to generate word cloud.")
```

```python
# ANS 6.


import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense


data = """
Basketball is a fast-paced sport played between two teams. Players score points by shooting the ball through the opponent’s basket.
It requires agility, teamwork, strategy, and skill. Dribbling, passing, and shooting are core techniques. 
Great players often practice for hours to refine their craft. Defense is just as important as offense.
The game teaches discipline, coordination, and leadership. Whether played on streets or in stadiums, basketball unites people around the world.
It inspires passion, competition, and resilience. Young athletes dream of joining the NBA or representing their countries in international tournaments.
"""

print("ANS 6.")
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
total_words = len(tokenizer.word_index) + 1

print("\n(i) Total unique words (vocab size):", total_words)
print("    Word Index:\n", tokenizer.word_index)


input_sequences = []
words = data.lower().split()

for i in range(1, len(words)):
    n_gram_sequence = words[:i+1]
    token_list = tokenizer.texts_to_sequences([" ".join(n_gram_sequence)])[0]
    input_sequences.append(token_list)

print("\n(ii)Sample tokenized sequences (first 5):")
for seq in input_sequences[:5]:
    print(seq)


max_seq_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')



xs = np.array(input_sequences[:, :-1])
ys = tf.keras.utils.to_categorical(input_sequences[:, -1], num_classes=total_words)



model = Sequential()
model.add(Embedding(total_words, 10, input_length=max_seq_len-1))
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print("\n(iii) Training model... ")
model.fit(xs, ys, epochs=200, verbose=0)
print("Model training complete!")


def generate_text(seed_text, next_words=20):
    print(f"\n Starting seed text: \"{seed_text}\"")
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)[0]

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += " " + word
                break
    return seed_text


generated_text = generate_text("basketball", next_words=30)
print("\nGenerated Text:\n", generated_text)
```
