Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data, clean and process it, perform statistical analysis, and generate visual insights to help the business make informed decisions.

Part I: Randomized Sales Data Generation (NumPy)

Initialize Random Seed

Generate Sales Data

Create a NumPy array of random sales data for 12 months (shape = (12, 4)), where each column represents a product category: Electronics, Clothing, Home & Kitchen, Sports.
Sales values should be between 1000 and 5000 units per month.
The range should be randomized using the seed.
Convert to DataFrame

Convert the cleaned NumPy array into a Pandas DataFrame with columns: Electronics, Clothing, Home & Kitchen, Sports and index: Months from Jan to Dec.

```python
import numpy as np
import pandas as pd

roll_number = 102317290
np.random.seed(roll_number)

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

sales_data = np.random.randint(1000, 5001, size=(12, 4))
df = pd.DataFrame(sales_data, columns=categories, index=months)
print(df)
```

Part II: Data Manipulation and Analysis (Pandas)

Display the first 5 rows and summary statistics of the DataFrame.
Calculate the total sales per category and total sales per month.
Calculate the average sales growth between consecutive months for each category.
Add the following columns to the DataFrame:
Total Sales: Sum of all categories per month.
Growth Rate: Percentage change in Total Sales from the previous month.
Apply a randomized discount:
If roll number is even, apply a 10% discount to the Electronics category.
If odd, apply a 15% discount to the Clothing category.

```python
print(df.head())
print(df.describe())

df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
print(df)

if roll_number % 2 == 0:
    df["Electronics"] *= 0.9
else:
    df["Clothing"] *= 0.85

print(df)
```

Part III: Visualizations (Matplotlib and Seaborn)

Plot monthly sales trends for each category using line plots.
Create box plots to show the sales distribution for each category.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df[categories].plot(kind='line', figsize=(10, 5), marker='o')
plt.title("Monthly Sales Trends")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

Q2. For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])

i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row.
iii. Mean, Median and Standard Deviation of flattened array, of each column, and each row

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_values = np.abs(array)
print(abs_values)

flattened = array.flatten()
print(np.percentile(flattened, [25, 50, 75]))
print(np.percentile(array, [25, 50, 75], axis=0))
print(np.percentile(array, [25, 50, 75], axis=1))

print(np.mean(flattened), np.median(flattened), np.std(flattened))
print(np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0))
print(np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1))
```

Q3. For the array: a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

i. Find floor, ceiling and truncated value, rounded values

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print(np.floor(a))
print(np.ceil(a))
print(np.trunc(a))
print(np.round(a))
```

Q4. Write a Python program to swap two elements in a list using a temporary variable.

```python
lst = [1, 2, 3, 4, 5]
i, j = 1, 3

temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

print(lst)
```

Q5. Write a Python program to swap two elements in a set by converting it to a list first.

```python
s = {10, 20, 30, 40, 50}
lst = list(s)
i, j = 1, 3

lst[i], lst[j] = lst[j], lst[i]
s = set(lst)

print(s)
```
